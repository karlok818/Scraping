```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

soup = BeautifulSoup(open("/Users/sookarlok/Downloads/network.html"), "html.parser") #'html.parser' if lxml not installed
regions = ['johor', 'kedah', 'kelantan', 'kuala-lumpur', 'melaka', 'negeri-sembilan', 'pahang', 'perak', 'perlis', 'pulau-pinang', 'putrajaya', 'sabah', 'sarawak', 'selangor', 'terengganu']
table = []

# for region in soup.select('.s-regional-analysis__region'):
#     regions.append(region["id"])
    
# print(regions)
    
for region in regions:

    res = soup.find_all('div', {'id':region})

    aval = res[0].find_all('div', {'data-chart-name':'4g-availability'})
    avalname = aval[0].find('span', {'class':'js-metric-name'}).text

    upload = res[0].find_all('div', {'data-chart-name':'upload-speed'})
    uploadname = upload[0].find('span', {'class':'js-metric-name'}).text

    download = res[0].find_all('div', {'data-chart-name':'download-speed'})
    downloadname = download[0].find('span', {'class':'js-metric-name'}).text
    
    video = res[0].find_all('div', {'data-chart-name':'video-experience'})
    videoname = video[0].find('span', {'class':'js-metric-name'}).text
    
    games = res[0].find_all('div', {'data-chart-name':'games-experience'})
    gamesname = games[0].find('span', {'class':'js-metric-name'}).text
    
    voice = res[0].find_all('div', {'data-chart-name':'voice-experience'})
    voicename = voice[0].find('span', {'class':'js-metric-name'}).text

    companies = [i.text for i in aval[0].find_all('span', class_='c-ru-graph__label')]

    row1 = [i.text for i in aval[0].find_all('span', class_='c-ru-graph__number')]
    row2 = [i.text for i in upload[0].find_all('span', class_='c-ru-graph__number')]
    row3 = [i.text for i in download[0].find_all('span', class_='c-ru-graph__number')]
    row4 = [i.text for i in video[0].find_all('span', class_='c-ru-graph__number')]
    row5 = [i.text for i in games[0].find_all('span', class_='c-ru-graph__number')]
    row6 = [i.text for i in voice[0].find_all('span', class_='c-ru-graph__number')]    

    df = pd.DataFrame({avalname:row1,
                       uploadname:row2,
                       downloadname:row3,
                       videoname:row4,
                       gamesname:row5,
                       voicename:row6
                      })

    df.index = companies
    df = df.T
    df['id'] = region
    table.append(df)
    
table = pd.concat(table)
table.to_csv('regional_analysis2021.csv')
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

soup = BeautifulSoup(open("/Users/sookarlok/Downloads/network2020.html"), "html.parser") #'html.parser' if lxml not installed
regions = ['video_experience', 'games_experience', 'voice_experience', 'download_speed', 'upload_speed', '4g_availability']
table = []

# for region in soup.select('.s-regional-analysis__region'):
#     regions.append(region["id"])
    
# print(regions)
    
for region in regions:

    res = soup.find_all('div', {'id':'video_experience'})
    provider=res[0].find_all('div', class_='c-ru-graph__rect')
    providername = [provider[i].find('span', class_='c-ru-graph__label').text for i in range(0,5)]
    provider=res[0].find_all('div', class_='c-ru-graph__rect')
    prodvalue = [provider[i].find_next_sibling('span').find('span', class_='c-ru-graph__number').text for i in range(0,5)]
    
    df = pd.DataFrame({
                        'provider':providername,
                        'value':prodvalue,     
                      })
    df['type'] = region
    table.append(df)
    
table = pd.concat(table)
table.to_csv('national_analysis2020.csv')
```
